We are told to use this equation: $$\frac{dP}{dz}=-\gamma$$
Integration:
$$dP=-\gamma dz$$
$$\int dP=-\int\gamma dz$$
$$P=-\int\gamma dz+C$$

The $\int\gamma dz$ term on the right side is essentially "the area under the curve of a $\gamma$ versus $z$ graph," which looks like this:

```julia
using Plots;
x=0:10:100;
y=[70,76,84,91,97,102,107,110,112,114,115];
ps = plot(x,y,xlim=(0,100),ylim=(60,120),legend=false,xlabel="specific weight (lb/ft^3)",ylabel="h (ft)")
```

We can use the trapezoid rule to approximate the area. Let $I$ be the integral and the partition width be $10 ft$.
$$I\approx\sum_{i=1}^{n}\frac{(h_i+h_{i-1})}{2}(10\ ft)=(5\ ft)\sum_{i=1}^{n}(h_i+h_{i-1})$$

Now we plug in our values for each point. As an example, for the depth $10\ ft$, $n=1$, and the pressure will be:
$$I_1\approx(5\ ft)(76+70\ lb/ft^3)=730\ lb/ft^2$$
Similarly, for $n=2$, we have:
$$I_2\approx(5\ ft)((76+70)+(84+76)\ lb/ft^3)=1530\ lb/ft^2$$

We can then plot pressure versus depth.

```julia
xf=[0:10:100];
yf=[0,730,1530,2405,3345,4340,5385,6470,7580,8710,9855];
ps = plot(xf,yf,xlim=(0,100),ylim=(0,10500),legend=false,xlabel="Depth Below Surface, h (ft))",ylabel="Pressure, P (lb/ft^2)")
```

Note that we found the gauge pressure of the liquid. The absolute pressure would account for the atmospheric pressure. The atmospheric pressure is also the value of the constant $C$ in the integrated expression from before: $P=-\int\gamma dz+C=-\int\gamma dz+P_{atm}$
